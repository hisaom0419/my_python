```python
#ライブラリーのインポート
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set the styles to Seaborn
sns.set()
# Import the KMeans module so we can perform k-means clustering with sklearn
from sklearn.cluster import KMeans
```

```python
#データの読み込み（入力待ちとif文の練習）
print('ファイル名を入力してください。')
file_name = input()

if('xls' in file_name):
  df = pd.read_excel(file_name, index_col=0)

elif('csv' in file_name):
  df = pd.read_csv(file_name, index_col=0)

else:
  print('incorrect file format')
```

```python
#表示
#df
```

```python
#相関を出す
corr_cond = df.corr()
corr_gene = df.transpose().corr()

#表示
#corr_gene
```

```python
sns_plot = sns.clustermap(df, method='ward', metric='euclidean', cmap="plasma")
plt.setp(sns_plot.ax_heatmap.get_yticklabels(), rotation=0)
plt.setp(sns_plot.ax_heatmap.get_xticklabels(), rotation=30)

plt.show()
```

```python
sns_plot = sns.clustermap(df, method='average', cmap='viridis')
plt.setp(sns_plot.ax_heatmap.get_yticklabels(), rotation=0)
plt.setp(sns_plot.ax_heatmap.get_xticklabels(), rotation=30)

plt.show()
```
